```python
import random

class GameOfLife:
    def __init__(self, grid):
        self.grid = grid
        self.grid_size = len(grid)

    @classmethod
    def create_random_grid(cls, row_size, col_size, initial_alive): #TODO: remove this method
        grid = [[False] * col_size for _ in range(row_size)]
        alive_cells = random.sample(range(row_size * col_size), initial_alive)
        for cell in alive_cells:
            row = cell // col_size
            col = cell % col_size
            grid[row][col] = True
        return cls(grid)
    @classmethod
    def create_custom_grid(cls, grid): #TODO: remove this method
        return cls(grid)

    def _get_neighbors(self, row, col):
        neighbors = []
        for i in range(-1, 2):
            for j in range(-1, 2):
                if i == 0 and j == 0:
                    continue
                new_row = row + i
                new_col = col + j
                if 0 <= new_row < len(self.grid) and 0 <= new_col < len(self.grid[0]):
                    neighbors.append(self.grid[new_row][new_col])
        return neighbors

    def evolve(self):
        actual_grid = self.get_grid() #self.grid
        new_grd = [[False for _ in range(len(actual_grid[0]))] for _ in range(len(actual_grid))]
        for row_index in range(len(actual_grid)):
            for col_index in range(len(actual_grid[0])):
                cell = actual_grid[row_index][col_index]
                cell_neighbors = self.__get_neighbors(row_index, col_index)
                live_neighbors = cell_neighbors.count(True)
                if live_neighbors == 3 or (live_neighbors == 2 and cell):
                    new_grd[row_index][col_index] = True
                else:
                    new_grd[row_index][col_index] = False
        self.grid = new_grd #TODO: reTURN ONLY NEW_GRID
        return print(f"The new grid has been created {new_grd}")

    def get_grid(self): #TODO: remove this method
        return self.grid
    
    def print_grid(self):
        for row in self.grid:
            print("".join(["[o]" if cell else "[x]" for cell in row]), end="\n")


# Main function for testing
def main():
    # Création d'une grille aléatoire de taille 5x5 avec 7 cellules vivantes
    random_game = GameOfLife.create_random_grid(4, 4, 5)
    
    print("Grille initiale :")
    print(random_game.print_grid())

    # Évolution de la grille et affichage à chaque étape
    num_steps = 5
    for step in range(num_steps):
        print("\nÉtape", step + 1)
        random_game.evolve()
        print(random_game.print_grid())

if __name__ == "__main__":
    main()
```
